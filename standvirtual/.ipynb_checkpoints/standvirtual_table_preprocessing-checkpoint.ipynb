```python
def read_shapefile(sf):
    """
    Read a shapefile into a Pandas dataframe with a 'coords' 
    column holding the geometry information. This uses the pyshp
    package
    """
    fields = [x[0] for x in sf.fields][1:]
    records = sf.records()
    shps = [s.points for s in sf.shapes()]
    df = pd.DataFrame(columns=fields, data=records)
    df = df.assign(coords=shps)
    return df

def plot_map(df, figsize=[5,5], cvals=None, label=''):
    if np.all(cvals == None):
        if len(df) <= 20:
            cmap = cm.get_cmap('tab20')            
        else:
            cmap = cm.get_cmap('jet')
        colors = cmap(np.linspace(0, 1, len(df)))
    else:
        cmap = cm.get_cmap('Blues')
        colors = cmap((cvals-cvals.min())/(cvals.max()-cvals.min()))
            
    fig, ax = plt.subplots(figsize=figsize)
    plt.axis('equal')
    k = 0
    for i, row in df.iterrows():
        name = row['NAME_1']
        X = np.array(row['coords'])
        plt.plot(X[:,0], X[:,1], 'k.')
        plt.fill(X[:,0], X[:,1], color=colors[k])
    
        x0 = np.mean(X[:,0])
        y0 = np.mean(X[:,1])
        plt.text(x0, y0, name, fontsize=15, ha='center', va='center', color='k')
        k += 1
    plt.axis('off')
    cax = fig.add_axes([0.7, 0.2, 0.15, 0.01])
    cbar = plt.colorbar(cm.ScalarMappable(norm=mpl.colors.Normalize(vmin=cvals.min(), vmax=cvals.max()), cmap=cmap),
                        ax=ax, fraction=0.1, orientation='horizontal', cax=cax)
    cbar.ax.set_xlabel(label)
    
    return fig
```

### Prepare data

```python
cols2use = [
    'fuel_type',
    'mileage',
    'power',
    'region',
    'reg_year',
    'garanty',
    'price_neg',
    'price',
    'advertiser',
    'brand',
    'model',
    'cylinder',
    'segment',
    'color',
    'metallic',
    'gear_type',
    'gears_n',
    'doors_n',
    'capacity',
    'air_conditioning',
    'vehicle_condition',
]

dfX = dfX[cols2use]

for column in dfX.columns:
 if dfX[column].dtype==object:
    dummyCols = pd.get_dummies(dfX[column])
    dummyCols.columns = column+'_'+dummyCols.columns.values
    dfX = dfX.join(dummyCols)
    del dfX[column]
    
dfX = dfX.loc[~dfX.isna().any(axis=1)]
```

```python
X = dfX.drop(columns='price').values
y = dfX['price'].values
```

## Ridge

```python
clf_rdg = RidgeCV(alphas=[1e-3, 1e-2, 1e-1, 1], cv=5).fit(X, y)
clf_rdg_score = clf_rdg.score(X, y)
print('Accuracy = %0.2f'%(clf_rdg_score))
```

## Lasso

```python
clf_lss = LassoLarsCV(cv=5, eps=1).fit(X, y)
clf_lss_score = clf_lss.score(X,y)

print('Accuracy = %0.2f'%(clf_lss_score))
```

## ElasticNet

```python
clf_ela = ElasticNetCV(cv=5).fit(X, y)
clf_ela_score = clf_ela.score(X,y)

print('Accuracy = %0.2f'%(clf_ela_score))
```

## T-SNE

```python
tsne_perp = [10,30,50]
tsne_dim = 2

X_tsne = np.zeros((len(X),tsne_dim,len(tsne_perp)))
for i in range(len(tsne_perp)):
    X_tsne[:,:,i] = TSNE(n_components=tsne_dim,
                         n_jobs=-1,
                         perplexity=tsne_perp[i]).fit_transform(X)
```

```python
fig, ax = plt.subplots(1,len(tsne_perp),
                       figsize=[7*len(tsne_perp),7])
for i in range(len(tsne_perp)):
    ax[i].scatter(X_tsne[:,0,i],X_tsne[:,1,i])
    ax[i].axis('equal')
    ax[i].set_title('Perplexity = %d'%(tsne_perp[i]))
plt.show()
```

```python
tsne_perp = 50
tsne_dim = 2


X_tsne = TSNE(n_components=tsne_dim, n_jobs=-1, perplexity=tsne_perp).fit_transform(X)
```

```python
fig, ax = plt.subplots(figsize=[7,7])
plt.scatter(X_tsne[:,0],X_tsne[:,1])
plt.axis('equal')
plt.show()
```

```python
ks = np.arange(2,100)
gmm_AIC = np.zeros(len(ks))

for i,k in enumerate(ks):
    gmm = mixture.GaussianMixture(n_components=k, covariance_type='full').fit(X_tsne)
    gmm_AIC[i] = gmm.aic(X_tsne)
```

```python
fig, ax = plt.subplots()
plt.plot(ks,gmm_AIC,'o-')
plt.show()
```

```python
k_final = 8
gmm = mixture.GaussianMixture(n_components=k_final, covariance_type='full').fit(X_tsne)
gmm_labels = gmm.predict(X_tsne)

fig, ax = plt.subplots(figsize=[7,7])
plt.scatter(X_tsne[:,0],X_tsne[:,1], c=gmm_labels)
plt.axis('equal')
plt.show()
```
